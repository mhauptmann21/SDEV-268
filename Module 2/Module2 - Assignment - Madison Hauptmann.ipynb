Exercise 2: Design principles and patterns

1. Define the key characteristics of Design Patterns and describe each.

    The key characteristics of design patterns are reusability, standardization, efficiency, and flexibility. Reusability refers to the idea of patterns being used in different projects and problems, saving time and effort in solving similar issues. Standardization means that developers will be able to understand the patterns, because they are widely used and provide a shared language. When a dveloper uses these design patterns, problems can be solved faster and more effiecent, because there is already a solution to the recurring problem. Patterns are similar to templates in which they can be adapted to various scenarios, leading to a lot of flexibility.

2. What are the critical reasons for learning design patterns?

    It is important to learn design patterns because they offer solutions that have already been tested, are simpler to comprehend and update, and saves development time.

3. What are the three types of software design patterns?

    The three types of software design patterns are Creational Design Pattern, Structural Design Pattern, and Behaviorial Design Pattern.

Exercise 3: Types of Software Design Patterns

1. Structural

    Adapter Design Pattern - Acts as a bridge between two incompatible interfaces, allowing them to work together.

    Bridge Design Pattern - Allows the developer to seperate the abstraction from the implementation.

2. Creational

    Factory Method - Defines an interface for creating objects but lets subclasses decide which object to instantiate.

    Prototype Design Pattern - Required when object creation is a time-consuming, and costly operation, so we create objects with the existing object itself by copying the existing ones.

3. Behavior

    Chain of Responsibility - Allows an object to pass a request along a chain of handlers. Each handler in the chain decides to either process the request or to pass it along to the next handler.

    State Design Pattern - Allows an object to change its behavior when its internal state changes.

Exercise 4: Observer Design Pattern

1. Observe Design Pattern - Establishes a one-to-many dependency between objects. When the subject changes its state, all of its observers are automatically notified and updated.

2. Disscuss the situations in wich using the Observer Design Pattern is appropriate.

    The Observer Design Pattern can be used for Social Media Notifications, Stock Market Apps, Event Listeners in GUIs, Weather Monitoring Systems, etc.



Exercise 5: Anti-Patterns

1. Explain the concept of an anti-pattern.

    An anti-pattern is the opposite side of the design pattern. Anti-patterns affect your software badly and it adds "technical debt" - more work for you, too. 

2. List and explain the six common anti-patterns in detail.

    Spaghetti Code - This is one of the most common types of anti-pattern. This type of code often results from poor planning, lack of programming style rules, or insufficient experience among software engineers. Spaghetti code is characterized by its complex and tangles control structure.

    Golden Hammer - This anti-pattern consists of developers becoming overly reliant on a familar tool or technique, applying it to every problem, regarless of its suitability. This can lead to suboptimal solutions, increased complexity, and reduced adaptability.

    Boat Anchor - This refers to when developers leave unused code or components in the system, contributing to unnecessary complexity and potential maintenance challenges. This makes the code harder to understand and maintain. They can obscure the functionality of the codebase, leading to confusion and errors.

    Dead Code - This refers to sections of code within a program that are never executed furing runtime, and have no impact on the program's output or behavior. Dead Code can include instructions that can never be executed, computations whose results are never used, or code that is unreachable due to conditiona; statements that are always false.

    God Object and God Class - This anti-pattern refers to a class that holds too much responsibility in a system. It violates key software design principles, such as Single Responsibility Principle (SRP), which states that a class should only have one reason to change. God Objects tend to grow uncontrollable data and methods that should logically belong to multiple small, specialized classes.

    Copy and Paste Programming - Copy and Paste Programming is the practice of reusing code by copying and pasting it, often leading to maintenance challenges and potential bugs in software development. This practice can arise from time constraints, lack of understanding, or the desire to quickly replicate functionality.
