Exercise 1 - Augmenting a data structure

17.1-5 Solution:
    1. Use the order-statistic routine to compute r = OS-RANK(x). This takes O(log n) time.
    2. Compute the desired rank ( t = r + i ). If ( t > n ), then the ith successor does not exist (report NIL).
    3. Return OS-SELECT(root, t). OS-SELECT runs in O(log n).

    Total Time: O(log n) + O(log n) = O(log n)

17.1-6 Solution:

    Insertion:
        1. When you create a new node z (a singleton), set r[z] = 1.
        2. When you search down from the root to find the insertion position for key k, do the usual BST descent. For every node y visited on the path:
            - If the descent goes to ( left(y) ) (the new node will be inserted into y's left subtree), increment r[y] by 1 (because left(y) gains one node). 
            - If the descent goes ( right(y) ), do not change r[y].
        3. Insert z as usual (leaf), with r[z] = 1.
        4. If the tree is a balanced tree and rotations occur to restore balance, update the r fields for only the nodes involved in each rotation (two or three nodes), using the formulas:
            - rank[y]new = rank[x]old + rank[y]old
        5. Rotations are O(1) per rotation, so maintenance cost remains O(log n) overall.
            
            - rank[y] = |left(y)| + 1
    
    Deletion:
        1. When locating the node to delete, while traversing from root to the node z to be removed:
            - If you go to the left(y) (the node to be removed is in y's left subtree), decrement r[y] by 1.
            - If you go to right(y), there is no change.
        2. Perform the usual BST deletion cases (leaf, one child, two children). If you replace/delete/internally transplant nodes, update the r fields of the directly affected nodes (local recalculation).
        3. As with insertion, if rotations occur to restore balancing invariants, apply the rotation update formulas below to the nodes involve:
            - rank[y]new = rank[x]old + rank[y]old

Exercise 2: Data Structures for Disjoint Sets

19.3-3:

    Adversarial Sequence:
        1. Perform n MAKE-SET operations to create singleton sets
            - {1}, {2},....,{n}
        2. Use unions in a way that builds one tree whose height is Θ(log n) (union by rank produces trees whose height is at most [log n]) For example:
            - Pair up and union adjacent singletons to form [n/2] sets of size 2 (ranks become 1)
            - Pair those to form size-4 sets, and so on -- perform unions in the standard "combine equal sizes" manner until you have one set of size n. This requires about n - 1 UNION operations.
        3. Now perform the remaining operations as repeated FIND-SET calls on a deepest leaf (a node at maximal depth). Because there is no path compression, each FIND-SET on that leaf walks up the tree to the root paying Θ(log n) time (tree height is Θ(logn n)).
        4. Choose the total count m so that after n MAKE-SETs and the Θ(n) UNIONs used to assemble the tree, you have many FINDs remaining. Each such FIND cost Θ(log n).

        Building the tree: Θ(n) work (UNIONs each are O(1) for union-by-rank ignoring recursive finds). Each of the remaining FINDs costs Θ(log n). If the number of remaining operations is Θ(m), total cost is Θ(m log n). So the sequence achieves Ω(m logn n).

Exercise 3: Elementry Graph Algorithms

20.2-5:

    1. BFS explores vertices in layers by increasing distance from the source s. Let δ(s,u) denote the shortest-path distance from s to u.
    2. BFS intialization sets s.d = 0, others = ∞, and uses a queue. When BFS discovers a new vertex v, it sets v.d = u.d + 1 where u is the vertex currently being processed that v.
    3. Key invariant: when BFS dequeues a vertex with distance k, all vertices at distance < k have already been discovered and dequeued. Vertices at distance k may be discovered in some order, but they are all discovered before any vertex at distance k + 1.
    4. By induction on distance k:
        - Base k = 0: s.d = 0 independent of adjacency order.
        - Inductive step: assume every vertex with δ(s,v) ≤ k gets v.d = δ(s,v) indpendent of adjacency ordering. Consider a vertex u with δ(s,u) = k + 1. there exists a shortest path s⇝⋯⇝w⇝u where δ(s,w) = k. By the induction hypothesis, w is discovered and enqueued with w.d = k. When w is dequeued (which happens before any vertex of distance k + 1 is dequeued), BFS inspects w's adjacency list and will sicover u ( or some vertex on another shortest path to u), assiging u.d = k + 1. Order inside adjacency lists may change which neighbor of a distance-k vertex is discovered first, but it cannot change the fact that some neighbor at distance k gets to discover u and thus u recieves k + 1.
    5. Therefore u.d is forced to equal δ(s,u) and is independent of adjacency ordering.

    Why the BFS tree can still depend on adjacency ordering:
        - The parent pointers (which neighbor first discovered u) can differ if multiple neighbors of u are at distance k. That choice depends on adjacency list order; hence the shape of BFS tree may change while all d-values remain the unique shortest distances.
