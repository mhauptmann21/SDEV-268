8.3-4 (page 215), Suppose that COUNTING-SORT is used as a stable sort within RADIX-SORT. If RADIX-SORT calls COUNTING-SORT d times, then since each call of COUNTING-SORT makes two passes over the data (lines 4-5 and 11-13), altogether, 2d passes over the data occur. Describe how to reduce the total number of passes to d + 1.

To reduce the number of passes to d + 1, you must eliminate the seperate "counting" pass for all but the first digit. There needs to be enough information from the previous sort so that for subsequent digits COUNTING-SORT only needs the single placement pass. 

You can use two arrays and alternate their roles, causing the removal of the copy-back step. This then leaves exactly one full pass per digit after the first.

2 passes (first digit) + (d-1) passes (remaining digits) = d + 1
