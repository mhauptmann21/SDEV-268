Exercise 1 - How does Dynamic Programming work?

1. Dynamic Programming is a problem-solving technique used when a problem can be broken down into smaller overlapping subproblems. Instead of solving the same subploblems repeatedly, Dynamic Programming stores the results so they can be reused later.

2. You should Dynamic Programming when a problem has overlapping subproblems or optimal substructure. If the same subproblems are solved repeatedly, then it beneficial to use Dynamic programming. Also, if an optimal solution can be built from the solutions of smaller subproblems, then Dynamic Programming becomes very helpful.

3. Dynamic Programming has two main approaches: Top-Down and Bottum Up. The first method, Top-Down, uses recursion to write the solution, stores answers to subproblems so that they aren't recomputed, and it solves only the subproblems needed. The next method, Bottom-Up, uses loops to build a table iteratively from the smallest subproblems up to the final answer.

4. The advantages of Dynamic Programming incudle that it is very efficient and avoids repeated calculations, it reduces time complexity, it guarantees optimal solutions, and it transforms recursive problems into iterative, stable ones.

Exercise 2 - Tabulation vs Memoization

int dp[MAXN];
// base case
int dp[0] = 1;
for (int i = 1; i <= n; i++)
{
    dp[i] = dp[i-1] * i;
}


Bottom-Up (Tabulation)

    - it starts from the base case (dp[0] = 1)
    - it fills up the table iteratively from small to large ( 1 -> n )
    - no recursion is used in the problem



int dp[MAXN]
// return fact x!
int solve(int x)
{
    if (x==0)
        return 1;
    if (dp[x]!=-1)
        return dp[x];
    return (dp[x] = x * solve(x-1));
}


Top-Down (Memoization)

    - uses recursion in the problem ( solve(x-1) )
    - it only computes needed states
    - the results are stored in ( dp[x] ) to avoid recomputation

Exercise 3 - Coin CHanging

Greedy Algorithm

To make change for n cents using the fewest coins:
    - use as many quarters as possible
    - then use as many dimes as possible
    - then use as many nickels as possible
    - use pennies for the remainder

numQuarters = n / 25
n = n % 25

numDimes = n / 10 
n = n % 10

numNickels = n / 5
n = n % 5

numPennies = n

To prove the greedy algorithm is optimal, we show that using the largest coin available never prevents us from getting a minimum-coin solution. 

    1. 25 > 10 + 10 + 1 + 1 + 1 + 1 + 1
    2. 10 > 5 + 5
    3. 5 > 1 + 1 + 1 + 1 + 1
    
    No combination of smaller coins can match the higher coin, using fewer coins.
